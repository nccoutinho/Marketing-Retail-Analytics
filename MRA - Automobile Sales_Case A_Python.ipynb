### Shapiro Wilk Test

```python
cols = ["QUANTITYORDERED", "PRICEEACH", "SALES", "DAYS_SINCE_LASTORDER", "MSRP"]

for col in cols:
    print(f'{col}')
    stat, p = shapiro(sales['QUANTITYORDERED'])
    print('Statistics=%.3f, p=%.3f' % (stat, p))

    alpha = 0.05
    if p > alpha:
        print('Sample looks Gaussian (fail to reject H0)')
    else:
        print('Sample does not look Gaussian (reject H0)')

    print('')
```

### Total Price

### RFM Customer Segmentation

```python
rfmTable = sales.groupby('CUSTOMERNAME').agg({'ORDERDATE': lambda x: (now - x.max()).days, 'ORDERNUMBER': lambda x: len(x), 'TotalPrice': lambda x: x.sum()})
rfmTable['ORDERDATE'] = rfmTable['ORDERDATE'].astype(int)
rfmTable.rename(columns={'ORDERDATE': 'recency', 
                         'ORDERNUMBER': 'frequency', 
                         'TotalPrice': 'monetary_value'}, inplace=True)

rfmTable
```

### Segmented RFM Table (4 Segments)

```python
quantiles = rfmTable.quantile(q=[0.25, 0.5, 0.75])
quantiles = quantiles.to_dict()
```

```python
segmented_rfm = rfmTable
```

```python
# Arguments (x = value, p = recency, monetary_value, frequency, d = quartiles dict)

def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 1
    
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
```

```python
segmented_rfm['r_quartile'] = segmented_rfm['recency'].apply(RScore, args=('recency',quantiles,))
segmented_rfm['f_quartile'] = segmented_rfm['frequency'].apply(FMScore, args=('frequency',quantiles,))
segmented_rfm['m_quartile'] = segmented_rfm['monetary_value'].apply(FMScore, args=('monetary_value',quantiles,))


segmented_rfm.head()
```

```python
segmented_rfm['RFMScore'] = segmented_rfm.r_quartile.map(str) + segmented_rfm.f_quartile.map(str) + segmented_rfm.m_quartile.map(str)
segmented_rfm.head()
```

The lowest recency is the most recent (4), highest frequency and monetary amounts (4).
